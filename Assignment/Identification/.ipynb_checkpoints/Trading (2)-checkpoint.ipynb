```python
import os
import datetime
import numpy as np
import scipy as sp
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Importing the dataset

df = pd.read_csv(os.getcwd() + '\\Dataset\\' + str(os.listdir('./Dataset')[0]), sep = ",", delimiter = ",",
                 header = 'infer', names = ['Time', 'Open', 'High', 'Low', 'Close', 'Volume', 'Volume MA'], 
                 skiprows = 1, encoding = 'ISO-8859-1')
```

```python
#Getting an idea of the structure

df.head()
```

```python
#Converting the first variable values into their corresponding timestamp

df['Time'] = pd.to_datetime(df['Time'], unit='s')
df.head()
```

```python
#Searching for non values 

df.isna().sum()
```

```python
#Converting the Time variable into the new index

df = df.set_index('Time')
df.head()
```

```python
#Creating a signals' dataframe to store information about those time instants during which a crossing shows off and the
#computed moving averages over two periods (10 and 20)

signals = pd.DataFrame(index=df.index)
signals['Signal'] = 0.0
signals['Short_mavg'] = df['Close'].rolling(10, 1).mean()
signals['Long_mavg'] = df['Close'].rolling(20, 1).mean()
signals['Signal'][10:] = np.where(signals['Short_mavg'][10:] > signals['Long_mavg'][10:], 1.0, 0.0)   
signals['Positions'] = signals['Signal'].diff()   
```

```python
#Plotting the moving averages curves

sns.set_style("whitegrid")
plt.figure(figsize=(15, 10))
plt.plot(df['Close'], color = 'gold', lw = 3)
plt.plot(signals['Short_mavg'][10:], lw = 3, color = 'grey')
plt.plot(signals['Long_mavg'][10:], lw = 3, color = 'darkblue')
plt.plot(signals.loc[signals.Positions == 1.0].index, signals.Short_mavg[signals.Positions == 1.0],
         '^', markersize=10, color='m')
plt.plot(signals.loc[signals.Positions == -1.0].index, signals.Short_mavg[signals.Positions == -1.0],
         'v', markersize=10, color='k')
plt.title('Close Moving Average', size = 20, pad = 20)
plt.xlabel('Time', size = 18, labelpad = 15)
plt.ylabel('Values', size = 18, labelpad = 15)
plt.xticks(size = 13, rotation = 45)
plt.yticks(size = 13)
plt.gca().legend(('Close curve', '10 period Moving AVG','20 period Moving AVG', 'Purchase', 'Sell'), loc = 'upper right', fontsize = 'xx-large',
                shadow = True, fancybox = True)
plt.show()
```

```python
# Creating a dataframe to store positions, namely: whenever the strategy concretizes in a purchase

positions = pd.DataFrame(index=df.index).fillna(0.0)
positions['Symbol'] = 100*signals['Signal']
```

```python
#Creating a portfolio dataframe 

portfolio = positions.multiply(df['Close'], axis = 'index')
```

```python
#Computing the difference between every position and the previous over a time reference

pos_diff = positions.diff()
```

```python
#Setting out a baseline quantity as starting capital amount

initial_capital = 100000.0
```

```python
#Updating the portfolio dataframe with derived variables

portfolio['Holdings'] = (positions.multiply(df['Close'], axis = 'index')).sum(axis = 1)
portfolio['Cash'] = initial_capital - (pos_diff.multiply(df['Close'], axis = 'index')).sum(axis = 1).cumsum()
portfolio['Total'] = portfolio['Cash'] + portfolio['Holdings']
portfolio['Returns'] = portfolio['Total'].pct_change()
```

```python
#Plotting the equity curve

sns.set_style("whitegrid")
fig = plt.figure(figsize=(15,10))
plt.plot(portfolio['total'], label = '__nolegend__', lw = 3, color = 'salmon')
plt.plot(portfolio.loc[signals.Positions == 1.0].index, portfolio.Total[signals.Positions == 1.0], '^', markersize = 10, color = 'm', label = "Purchase")
plt.plot(portfolio.loc[signals.Positions == -1.0].index, portfolio.Total[signals.Positions == -1.0], 'v', markersize = 10, color = 'k', label = 'Sell')
plt.title('Equity line', size = 20, pad = 20, fontweight = 'bold')
plt.xlabel('Indexed time', size = 18, labelpad = 15)
plt.ylabel('Total', size = 18, labelpad = 15)
plt.xticks(size = 13, rotation = 45)
plt.yticks(size = 13)
plt.legend(loc = 'upper right', fontsize = 'xx-large', shadow = True, fancybox = True)
plt.show()
```
